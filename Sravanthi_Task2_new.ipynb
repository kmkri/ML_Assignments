```python
#importing the required libraries from uciml repository
from ucimlrepo import fetch_ucirepo
import pandas as pd
```

```python
#fectching the Adult dataset from the repository
dataset = fetch_ucirepo(name = 'Adult')
dataset.variables
```

```python
#creating the dataframe
x = pd.DataFrame(dataset.data.features)
y = pd.DataFrame(dataset.data.targets)
df_Adult = pd.concat([x,y],axis = 1)
df_Adult
```
