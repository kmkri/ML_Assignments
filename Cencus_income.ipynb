```python
#Fetching UCI repositories 
from ucimlrepo import fetch_ucirepo, list_available_datasets
```

```python
#List of UCI repositories with ID
list_available_datasets()
```

```python
# fetch dataset 
adult = fetch_ucirepo(id=2) 
  
# data (as pandas dataframes) 
X = adult.data.features 
y = adult.data.targets 
  
# metadata 
print(adult.metadata) 
```

```python
# variable information 
print(adult.variables) 
```

```python
#Creating dataframe by joining features and target
Xdata = pd.DataFrame(X)
Ydata = pd.DataFrame(y, columns=['income'])
adultDataFrame = pd.concat([Xdata,Ydata], axis = 1)
adultDataFrame
```
